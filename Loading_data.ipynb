```python
import pandas as pd

# File paths
amazon_path = "Amazon.csv"
goog_path = "GOOG.csv"
nflx_path = "NFLX.csv"

# Read the CSV files
amazon_df = pd.read_csv(amazon_path)
goog_df = pd.read_csv(goog_path)
nflx_df = pd.read_csv(nflx_path)

# Function to display shape and info
def dataset_info(df, name):
    print(f"--- {name} Dataset ---")
    print("Shape:", df.shape)
    print("Info:")
    print(df.info())
    print("\n")

# Display info for each dataset
dataset_info(amazon_df, "Amazon")
dataset_info(goog_df, "Google")
dataset_info(nflx_df, "Netflix")
```
